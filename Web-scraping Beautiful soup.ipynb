### Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.¶

```python
page8 = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJQAcLUzaUk')
soup8 = BeautifulSoup(page8.content)
print (soup8.prettify())
```

```python
period = soup8.find_all('p', class_ = 'period-name')
period
```

```python
period_name = []
for i in period:
    period_name.append(i.text.replace('\n', ''))
period_name
```

```python
short_desc = soup8.find_all('p', class_ = 'short-desc')
short_desc
```

```python
short_description = []
for i in short_desc:
    short_description.append(i.text.replace('\n', ''))
short_description
```

```python
temp_h = soup8.find_all('p', class_ = 'temp temp-high')
temp_h
```

```python
temp_high = []
for i in temp_h:
    temp_high.append(i.text)
temp_high
```

```python
temp_l = soup8.find_all('p', class_ = 'temp temp-low')
temp_l
```

```python
temp_low = []
for i in temp_l:
    temp_low.append(i.text)
temp_low
```

```python
desc = soup8.find_all('div', class_ = 'col-sm-10 forecast-text')
desc
```

```python
description = []
for i in desc:
    description.append(i.text.replace('\n', ''))
description = description[:9]
```

```python
v = 0
v1 = 0
v2 = 0
total_len = len(temp_low) + len(temp_high)
temperature = []

while temp_low or temp_high:
    if (v%2 == 0 and len(temp_high) > len(temp_low)):
        temperature.append(temp_high.pop())
    elif (v%2 == 0 and len(temp_low) > len(temp_high)):
        temperature.append(temp_low.pop())
    elif (v%2 == 1 and len(temp_high) > len(temp_low)):
        temperature.append(temp_low.pop())
    else:
        temperature.append(temp_high.pop())
    v +=1

temperature.reverse()
temperature
```

```python
weather = pd.DataFrame({})
weather['Period'] = period_name
weather['Short Description'] = short_description
weather['Temperature'] = temperature
weather['Description'] = description
weather
```
